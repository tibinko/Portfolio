```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('ppd_data_3_truncated.csv')

# Print column names
print(df.columns)
df['deed_date'] = pd.to_datetime(df['deed_date'])

# Extract only the year from the 'deed_date' column
df['year'] = df['deed_date'].dt.year
# Check data types
print(df.dtypes)

# Create a scatterplot using Seaborn
sns.scatterplot(x='deed_date', y='price_paid', data=df)

# Add labels and title
plt.title('Terraced House Prices in Chingford')
plt.xlabel('Date')
plt.ylabel('Price Paid')

# Show the plot
plt.show()
```
